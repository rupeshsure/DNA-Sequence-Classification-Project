```python
Sequence="atatgaacgttgagactgccgctgagttatcagctgtgaacgacattctggcgtcta"
```

```python
sequences = list([ 'atatgaacgttgagactgccgctgagttatcagctgtgaacgacattctggcgtcta'])
dataset = {}
for i, seq in enumerate(sequences):
   nucleotides = list(seq)
   nucleotides = [x for x in nucleotides if x != '\t']
   #nucleotides.append(classes[i])
   dataset[i] = nucleotides
   print(dataset[0])
  
```

```python
import pandas as pd
dframe = pd.DataFrame(dataset)
print(dframe)
```

```python
dff = dframe.transpose()
dff
```

```python
new=[]
for name in dff.columns:
    new.append(dff[name].value_counts())
```
